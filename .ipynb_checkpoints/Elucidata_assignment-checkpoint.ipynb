# 2.1 What does the analysis say about the general behaviour of the different samples?

A PCA plot converts the correlations among all the cells into a 2-D graph. Here in our case we can clearly see 2 different clusters(not considering some OUT-LIERS). Also we can see for PC1 most values are between range (-100 , 100) and for PC2 the range for most of the data is (-50 , 50).
Also, PC1 has more importance as compared to PC2, so we can say that samples are mostly showing common behavior, as distance between clusters can be seen from PC2-axis and that too is not very great.

# 2.3 Are the neuroendocrine tumors clearly separable from the adenocarcinoma tumors?Overlay the information from metadata column ‘histological_type_other’ on top of PCA plotand check if neuroendocrine tumors are separating out.

![title](2.png)

Yes,neuroendocrine tumors clearly separable from the adenocarcinoma tumors.(due to so many types inside ‘histological_type_other’ graph is not clearly captured, but is read by me thoroughly)(Using phantasus)

# 2.4 What can be said about the variance of the PCA?

The explained variance tells you how much information (variance) can be attributed to each of the principal components.You can see that the first principal component contains 13.9% of the variance and the second principal component contains 11.1% of the variance. Together, the two components can explain 25% of the information about variance in the data. So, In our case, we can say that both the clusters are almost equal in variance information, although PC1 has slightly more important when comes to have information about the variance of the data.

```python
# Final dataframe that contains gene expression values for these 25 genes for pancreatic adenocarcinoma(24 genes as 1 gene had NaN value and was removed during cleaning process)
genes = ['IFIT1','IFI44','IFIT3','MX2','OAS1','OAS3','BST2','IFITM1','MX1','STAT1','IFI27','CXCL10','IFI16','IFI30','IFIH1','IFIT2','IFITM2','IRF1','IRF9','IRGM','ISG15','OAS2','PSME1','SOCS1','STAT2']
viz = data.loc[data['rid']. isin (genes)]
viz
```

```python
# Dataframe which is used in GSVA algorithm.
viz_gsva = viz.drop(['rid'], axis = 1) 
viz_gsva
```

```python
# Getting attributes(column names)
cl =[]
for col in viz.columns: 
    cl.append(col)
```

```python
cl.remove('rid')
cl
```

```python
viz.plot(x="rid", y=cl, kind="bar", legend=False)
```
